```python
# gürültü ekleme = noise
import cv2
from matplotlib import image
import numpy as np
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings("ignore")

img = cv2.imread("messi5.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) / 255 # gürültü ekleyebilmek için 255'e bölerek scaling


# gürültü ekleme
def gausianNoise(image):
    row, col, ch = image.shape
    mean = 0
    var = 0.05
    sigma = var **0.5         # standart sapma

    gauss = np.random.normal(mean, sigma, (row, col, ch))
    gauss = gauss.reshape(row, col, ch)
    noisy = image + gauss

    return noisy     # oluşan gürültü 0-1 arasındadır

# tuz karabiber gürültüsü
def saltPepperNoisy(image):
    row, col, ch = image.shape
    s_vs_p =  0.05   # salt papper oranı
    amount = 0.004
    noisy = np.copy(image)     # orj,nal resmi bozmamak için kopyası ile çalışalım

    # salt gürültüsü
    num_salt = np.ceil(amount * image.size * s_vs_p)
    coords = [np.random.randint(0,i - 1, int(num_salt)) for i in image.shape]
    noisy[coords] = 1

    # papper gürültüsü
    num_papper = np.ceil(amount * image.size * (1-s_vs_p))
    coords = [np.random.randint(0,i - 1, int(num_papper)) for i in image.shape]
    noisy[coords] = 0
    # papper ve salt gürültü toplamı bir eder
    return noisy

noisy = gausianNoise(img)
salt_papper_img = saltPepperNoisy(img)

# görselleştirme
plt.figure(),plt.imshow(img),plt.axis("off"),plt.title("orijinal")
plt.figure(),plt.imshow(noisy),plt.axis("off"),plt.title("gürültülü bulanıklaştırma")
plt.figure(),plt.imshow(salt_papper_img),plt.axis("off"),plt.title("salt-papper bulanıklaştırma")


plt.show()

# NOT : gürültülerden bulanıklaştırmalardan kurtulunabilir
```

```python
# morfolojik işlemler

import cv2
import numpy as np
import matplotlib.pyplot as plt
from numpy.core.fromnumeric import size
from numpy.lib.function_base import gradient

# erozyon : görüntüdeki sınırların incelmesi 
# genişleme : görüntüdeki sınırların genişlemesi

# açılma : önce erozyon sonra genişleme yapılması
# kapatma : önce genişleme sonra erozyon
 
# Morfolojik gradyan : bir görüntünün genişlemesi ve erozyonu arasındaki farktır

img = cv2.imread("datai_team.jpg",0)
plt.figure(),plt.imshow(img, cmap="gray"),plt.axis("off"),plt.title("orijinal resim"),plt.show()

# erezyon  (kernel fotoğraf üzerinde gezinen kutucuk)
kernel = np.ones((5,5), dtype=np.uint8)
result = cv2.erode(img, kernel, iterations = 1)   
plt.figure(),plt.imshow(result, cmap= "gray"),plt.axis("off"),plt.title("erozyon"),plt.show()

# dilation = genişleme
result_2 = cv2.dilate(img, kernel, iterations=1)
plt.figure(),plt.imshow(result_2, cmap="gray"),plt.axis("off"),plt.title("dilation"),plt.show()

# açılma ile beyaz noise giderme

# gürültü oluşturalım
white_noise = np.random.randint(0,2, size = img.shape[:2])
white_noise = white_noise * 255
white_noise_img = white_noise +  img
plt.figure(),plt.imshow(white_noise_img, cmap="gray"),plt.axis("off"),plt.title("gürültü"),plt.show()

# açılma 
opening = cv2.morphologyEx(white_noise_img.astype(np.float32), cv2.MORPH_OPEN, kernel) 
plt.figure(), plt.imshow(opening, cmap="gray"), plt.axis("off"), plt.title("opening"), plt.show()


# kapatma ile siyah gürüşltüleri engelleme

# siyah gürültü oluşturma
blackNoise = np.random.randint(0,2, img.shape[:2])
blackNoise = blackNoise * (-255)

blackNoise_img = blackNoise + img
blackNoise_img[blackNoise_img <= -255] = 0
plt.figure(), plt.imshow(blackNoise_img, cmap="gray"),plt.axis("off"),plt.title("black noise"),plt.show()

# kapatma ile siyah gürültü giderme
closing = cv2.morphologyEx(blackNoise_img.astype(np.float32), cv2.MORPH_CLOSE, kernel)
plt.figure(),plt.imshow(closing, cmap="gray"),plt.axis("off"),plt.title("closing"),plt.show()


# morfolojik gradiyen 
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel) 
plt.figure(),plt.imshow(gradient,cmap="gray"),plt.axis("off"),plt.title("gradient"),plt.show()
```

```python
# kenar algılama

import matplotlib.pyplot as plt
import cv2
import numpy as np
from numpy.lib.function_base import median

img = cv2.imread("london.jpg",0)
plt.figure(), plt.imshow(img, cmap="gray"), plt.axis("off"), plt.title("siyahbeyaz")

# kenaralgılama fonksiyonu cv2.Canny
edges = cv2.Canny(image=img, threshold1 = 0, threshold2 = 255)
plt.figure(), plt.imshow(edges, cmap = "gray"), plt.axis("off"), plt.title("basit kenar algılama")

# threshold değerleri ile oynarsak
med_val = np.median(img)
print(med_val)

low = int( max(0, (1 - 0.33) * med_val))
high = int( min(255, (1 + 0.33 * med_val)))

edges = cv2.Canny(image=img, threshold1=low, threshold2 = high)
plt.figure(), plt.imshow(edges, cmap = "gray"), plt.axis("off"), plt.title("düzenlenmiş threshold")

"""
su dalgalarının yansımaları yüzünden kenarmış gibi algılanıyor, oradaki renk değişimlerini 
azalmak için blurlama yapılabilir 
"""
# bulur
blurred_img = cv2.blur(img, ksize=(7,7))

med_val = np.median(blurred_img)
low = int( max(0, (1 - 0.33) * med_val))
high = int( min(255, (1 + 0.33)))

edges = cv2.Canny(image = blurred_img, threshold1 = low, threshold2 = high)
plt.figure(), plt.imshow(edges, cmap = "gray"), plt.axis("off"), plt.title("Blur ile kenaralgılama")

# blurlamadaki kernel size ile oynanarak kenaralgılama düzenlenebilir 

plt.show()
```

```python
# köşe algılama

import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread("sudoku.jpg", 0)
img  = np.float32(img)
plt.figure(), plt.imshow(img, cmap="gray"), plt.axis("off"), plt.title("orijinal")

# harris corner detection
# dst = cv2.cornerHarris(image, blockSize = , ksize = , k =0.04 ) --- (resim ,komşuluk boyutu, ksize , k)
dst = cv2.cornerHarris(img, blockSize= 2, ksize = 3, k = 0.04)
plt.figure(), plt.imshow(dst), plt.axis("off"), plt.title("harris")

# resime açma uygulayarak köşe belirginleştirilebilir
dst = cv2.dilate(dst, None)
plt.figure(), plt.imshow(dst), plt.axis("off"), plt.title("harris açma")

# shi tomsai detection
corners = cv2.goodFeaturesToTrack(img, 120, 0.01, 10) # (resim , kçşe sayısı, quality_level, min_distance)
corners = np.int64(corners)

for i in corners:    
    x,y = i.ravel()    # konumu
    cv2.circle(img, (x,y), 3,  (125,125,125), cv2.FILLED)

plt.figure(), plt.imshow(img, cmap="gray"), plt.axis("off"), plt.title("shi tomsai")
plt.show()
```

```python
# kontur algılama  - contours

from typing import Counter
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("london.jpg", 0)
plt.figure(), plt.imshow(img, cmap = "gray"), plt.axis("off"), plt.title("orijinal")

contours, hierarche = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE) 
# (resim, kontur bulma yöntemi, yalnızca uç noktalarını bırakması)

externel_contours = np.zeros(img.shape)
internel_counters = np.zeros(img.shape)

for i in range(len(contours)):
    
    if hierarche[0][i][3] == -1 :   # exterenl 
        cv2.drawContours(externel_contours, contours, i, 255, -1)
    else: #internel
        cv2.drawContours(internel_counters, contours, i, 255, -1)

plt.figure(), plt.imshow(externel_contours, cmap = "gray"), plt.axis("off"), plt.title("externel")
plt.figure(), plt.imshow(internel_counters, cmap=  "gray"), plt.axis("off"), plt.title("internel")

plt.show()
```

```python
# renk ile nesne tespiti

from typing import Counter
import cv2
import numpy as np
import matplotlib.pyplot as plt
from collections import deque

# nesne merkezlerini tutma
buffer_size = 16
pts = deque(maxlen = buffer_size)

# mavi renk aralığı
blue_lower=np.array([100,150,0],np.uint8)
blue_upper=np.array([140,255,255],np.uint8)


# capture
cap = cv2.VideoCapture(0)
cap.set(3, 960)     # genişlik belirleme
cap.set(4, 480)     # yükseklik belirleme

while True:
    success, img_orijinal = cap.read()

    if success == True:  # doğru okundumu 

        # gürültüyü azaltmak için blurlama
        blured = cv2.GaussianBlur(img_orijinal, (11,11),0)  
        # HSV renk uzayına dönüştürme
        hsv = cv2.cvtColor(blured, cv2.COLOR_BGR2HSV)
        cv2.imshow("HSV image", hsv)

        # mavi için maske oluştur
        mask = cv2.inRange(hsv, blue_lower, blue_upper)
        cv2.imshow("mavi", mask)

        # maske etrafında gürültü temizleme (erozyon + genişleme)
        mask = cv2.erode(mask, None, iterations= 2)
        mask = cv2.dilate(mask, None, iterations= 2)
        cv2.imshow("temizlenmiş MASK", mask)

        # kontur
        (contours, _) = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        center = None

        if len(contours) > 0 :
            # en büyük kontur alma
            c = max(contours, key= cv2.contourArea)

            # boinding box
            rect  = cv2.minAreaRect(c)

            ((x,y),(width, height), rotation) = rect

            s = "x: {} , y: {}, height: {}, rotation: {} ".format(np.round(x),np.round(y),np.round(width),np.round(height))
            print(s)

            box = cv2.boxPoints(rect)
            box = np.int64(box)

            # moment  (görüntü merkezini bulmamızı sağlar)
            M = cv2.moments(c)
            center = (int(M["m10"] / M["m00"]), int(M["m01"] / M["m00"]))


            #kontur çizdirme
            cv2.drawContours(img_orijinal, [box], 0, (0,255,255), 2 )
            # merkeze nokta ekleme
            cv2.circle(img_orijinal, center, 5, (255,0,255), -1)
            # merkeze bir nokta daha ekleme
            cv2.circle(img_orijinal, center, 5, (255,2,255),-1)

            # bilgileri ekrana yazdırma
            cv2.putText(img_orijinal, s, (50,50), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0,0,0), 2)


        # deque ile nesne merkezinin hareketlerini çizdirme
        pts.appendleft(center)

        for i in range(1,len(pts)):
            if pts[i - 1] is None or pts[i] is None: continue

            cv2.line(img_orijinal, pts[i - 1], pts[i], (0,255,0), 3)

        cv2.imshow("orijinal tespit edilmiş", img_orijinal)

    if cv2.waitKey(1) & 0xFF == ord("q"): break

cap.relaise()
cv2.destroyAllWindows()
```

```python
# şablon eşleme - templete maching

import cv2
import matplotlib.pyplot as plt

# template matching: sablon esleme

img = cv2.imread("cat.jpg", 0)
print(img.shape)
template = cv2.imread("cat_face.jpg", 0)
print(template.shape)
h, w = template.shape

methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

for meth in methods:
    
    method = eval(meth) # 'cv2.TM_CCOEFF' -> cv2.TM_CCOEFF
    
    res = cv2.matchTemplate(img, template, method)
    print(res.shape)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    
    bottom_right = (top_left[0] + w, top_left[1] + h)
    
    cv2.rectangle(img, top_left, bottom_right, 255, 2)
    
    plt.figure()
    plt.subplot(121), plt.imshow(res, cmap = "gray")
    plt.title("Eşleşen Sonuç"), plt.axis("off")
    plt.subplot(122), plt.imshow(img, cmap = "gray")
    plt.title("Tespit edilen Sonuç"), plt.axis("off")
    plt.suptitle(meth)
    
    plt.show()
```

```python
# havza algoritması

import cv2
import matplotlib.pyplot as plt
import numpy as np

# içe aktar
coin = cv2.imread("coins.jpg")
plt.figure(), plt.imshow(coin), plt.axis("off")

# lpf: blurring
coin_blur = cv2.medianBlur(coin, 13)
plt.figure(), plt.imshow(coin_blur), plt.axis("off")

# grayscale
coin_gray = cv2.cvtColor(coin_blur, cv2.COLOR_BGR2GRAY)
plt.figure(), plt.imshow(coin_gray, cmap="gray"), plt.axis("off")

# binary threshold
ret, coin_thresh = cv2.threshold(coin_gray, 75, 255, cv2.THRESH_BINARY)
plt.figure(), plt.imshow(coin_thresh, cmap="gray"), plt.axis("off")

# kontur
# _, contours, hierarchy = cv2.findContours(coin_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
contours, hierarchy = cv2.findContours(coin_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(coin, contours,i,(0,255,0),10)
plt.figure(),plt.imshow(coin),plt.axis("off")

# watershed

# içe aktar
coin = cv2.imread("coins.jpg")
plt.figure(), plt.imshow(coin), plt.axis("off")

# lpf: blurring
coin_blur = cv2.medianBlur(coin, 13)
plt.figure(), plt.imshow(coin_blur), plt.axis("off")

# grayscale
coin_gray = cv2.cvtColor(coin_blur, cv2.COLOR_BGR2GRAY)
plt.figure(), plt.imshow(coin_gray, cmap="gray"), plt.axis("off")

# binary threshold
ret, coin_thresh = cv2.threshold(coin_gray, 65, 255, cv2.THRESH_BINARY)
plt.figure(), plt.imshow(coin_thresh, cmap="gray"), plt.axis("off")

# açılma
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(coin_thresh, cv2.MORPH_OPEN, kernel, iterations = 2)
plt.figure(), plt.imshow(opening, cmap="gray"), plt.axis("off")

# nesneler arası distance bulalım
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
plt.figure(), plt.imshow(dist_transform, cmap="gray"), plt.axis("off")

# resmi küçült
ret, sure_foreground = cv2.threshold(dist_transform, 0.4*np.max(dist_transform),255,0)
plt.figure(), plt.imshow(sure_foreground, cmap="gray"), plt.axis("off")

# arka plan için resmi büyült
sure_background = cv2.dilate(opening, kernel, iterations = 1)
sure_foreground = np.uint8(sure_foreground)
unknown = cv2.subtract(sure_background,sure_foreground)
plt.figure(), plt.imshow(unknown, cmap="gray"), plt.axis("off")

# bağlantı
ret, marker = cv2.connectedComponents(sure_foreground)
marker = marker + 1
marker[unknown == 255] = 0
plt.figure(), plt.imshow(marker, cmap="gray"), plt.axis("off")

# havza
marker = cv2.watershed(coin,marker)
plt.figure(), plt.imshow(marker, cmap="gray"), plt.axis("off")


# kontur
# _, contours, hierarchy = cv2.findContours(coin_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
contours, hierarchy = cv2.findContours(marker.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(coin, contours,i,(255,0,0),2)
plt.figure(),plt.imshow(coin),plt.axis("off")





plt.show()
  
```

```python
# yüz tanıma 

# yüz tanıma için haar özelliği tabanlı kademeli sınıflandırıcılar kullanılır

import cv2
import numpy as np
import matplotlib.pyplot as plt

# tek resimde - tek yüz
einstein = cv2.imread("einstein.jpg")
plt.figure(), plt.imshow(einstein), plt.axis("off"), plt.title("einstein")

# sınıflandırıcı
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

face_rect = face_cascade.detectMultiScale(einstein)

for (x,y,w,h) in face_rect:
    cv2.rectangle(einstein, (x,y),(x+w, y+h),(255,255,255),10)
plt.figure(), plt.imshow(einstein), plt.axis("off"), plt.title("einstein_tespit")

# tek resimde birden fazla yüz
barce = cv2.imread("barcelona.jpg",0)
plt.figure(), plt.imshow(barce,cmap ="gray"), plt.axis("off"), plt.title("barça")

# minNeighbors parametresini büyütmek tespit edilen yüzü algıalar
face_rect = face_cascade.detectMultiScale(barce, minNeighbors = 7)

for (x,y,w,h) in face_rect:
    cv2.rectangle(barce, (x,y), (x+w, y+h),(255,0,0),10)

plt.figure(), plt.imshow(barce, cmap = "gray"), plt.axis("off"), plt.title("barca detected")

plt.show()


# real time yüz tespit etme

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    if ret:
        
        face_rect = face_cascade.detectMultiScale(frame, minNeighbors = 5)
        
        for (x,y,w,h) in face_rect:
            cv2.rectangle(frame, (x,y),(x+w, y+h),(255,0,0), 10)
        cv2.imshow("face_detected", frame)

    if cv2.waitKey(1) & 0xFF == ord("q") : break

cap.relaise()
cv2.destroyAllWindows()
```

```python
# cascade oluşturma

# 1- cascade için resim kaydetme
import cv2
import os

files = os.listdir()
#print(files)

# resimleri içeren liste yapalım
img_path_list = []
for i in files:
    if i.endswith(".jpg"):
        img_path_list.append(i)

print(img_path_list)

# resimleri görselleştirme
for i in img_path_list: 
    img = cv2.imread(i)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    detector = cv2.CascadeClassifier("haarcascade_frontalcatface.xml")
    rects = detector.detectMultiScale(gray, scaleFactor = 1.045, minNeighbors = 2) # scaleFactor - ne kadar zoom yapacağı

    for x,y,w,h in rects:
        cv2.rectangle(img, (x,y), (x+w, y+h),(255,255,255),2)
        cv2.putText(img, "kedi", (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (255,255,255),2 )

    cv2.imshow("kedi", img)
    if cv2.waitKey(1) & 0xFF == ord("q"): continue


# 2-) cascade programı indir ( https://amin-ahmadi.com/cascade-trainer-gui/ ) 
#programdan custom cascade eğitilir ve xml dosyası alınır

# daha sonra cascadeclassfier ile bir kod  yazılarak tespit edme yapılıt


# parametrelerle oynanak için kod
import cv2

objectName = "Kalem Ucu"
frameWidth = 280
frameHeight = 360
color = (255,0,0)

cap = cv2.VideoCapture(0)
cap.set(3,frameWidth)
cap.set(4,frameHeight)

def empty(a):pass

# trackbar
cv2.namedWindow("Sonuc")
cv2.resizeWindow("Sonuc", frameWidth, frameHeight + 100)
cv2.createTrackbar("Scale","Sonuc",400,1000,empty)
cv2.createTrackbar("Neighbor","Sonuc",4,50,empty)

# cascade classifier
cascade = cv2.CascadeClassifier("cascade.xml")

while True:
    
    # read img
    success, img = cap.read()
    
    if success:
        # convert bgr2gray
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # detection parameters
        scaleVal = 1 + (cv2.getTrackbarPos("Scale","Sonuc")/1000)
        neighbor = cv2.getTrackbarPos("Neighbor","Sonuc")
        # detectiom
        rects = cascade.detectMultiScale(gray, scaleVal, neighbor)
    
        for (x,y,w,h) in rects:
            
            cv2.rectangle(img, (x,y),(x+w,y+h), color, 3)
            cv2.putText(img, objectName, (x, y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, color, 2)
            
        cv2.imshow("Sonuc", img)
    
    if cv2.waitKey(1) & 0xFF == ord("q"): break
    
    
    
    
    
    
    
```
